$T = \frac{a}{b}T^*$

$T_C = \frac{8a}{27b} = \frac{8}{27}\frac{T}{T^*}$

$T^* = \frac{8}{27}\frac{T}{T_C}$

with $\frac{T}{T_C} = [2, 1, 0.5]$

```python
P = lambda v,T: (8*T)/(3*v-1) - 3/v**2
P = lambda v,T: (8*T)/(27*(v-1)) - 1/v**2
P = lambda v,T: (8/27)*T/(v-1) - 1/v**2

v = np.linspace(1, 5, 1000)

rapp_T = [2, 1, 0.5]

fig, ax = plt.subplots(1,1, figsize=(5,5))

for i in range(3):
    ax.plot(v, P(v,rapp_T[i]))

ax.set_xlabel(r'$v^*$')
ax.set_ylabel(r'$P^*(v^*,T^*)$')
ax.set_title(r'Reduced unit Pressure')
ax.set_ylim(-1, 5)
ax.legend([r'$T/T_C=2$', r'$T/T_C=1$', r'$T/T_C=0.5$'])



```

```python
#f = lambda nl: (nl-ng)*(2-ng-nl) - (nl+ng)*(1-nl)*(1-ng)*np.log((nl*(1-ng))/(ng*(1-nl)))

def f(nl):
    global ng
    return (nl-ng)*(2-ng-nl) - (nl+ng)*(1-nl)*(1-ng)*np.log((nl*(1-ng))/(ng*(1-nl)))

def df(nl):
    global ng
    return -(2*nl + ng - 1)*(nl*(ng-1)*np.log(nl*(ng-1)/(ng*(nl-1)))+nl-ng)/nl

ngg = np.linspace(0.1, 0.33, 1000)

ind = np.where(ngg <= 0.99)[0]

nl = np.zeros(len(ind))

for i in range(len(ind)):
    ng = ngg[i]
    nl[i] = newton(f, 0.5, tol=1.e-3, maxiter=100, fprime=df)

```

```python
fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.plot(ngg[ind], nl, label='nl')
ax.set_xlabel(r'$n_g$')
ax.set_ylabel(r'$n_l$')
ax.set_title(r'$n_l(n_g)$')
```

# PROBLEM 6

In class, we have seen that the mean field approximation for the Ising model in zero
external field (h = 0) leads to the self-consistent relation for the magnetization

$m = \tanh (\beta zJm)$ (56)

where $J$ is the coupling constant and $z$ the coordination number. We have also seen that
the critical temperature is $k_BT_C = Jz$.

a) Plot the two curves in Eq.(56) for $T/T_c = [2, 1, 1/2]$ in the range $-1 <= m <= 1$ to
show explicitly that there is only 1 solution above and 3 solutions below the critical
temperature $T_C$.

```python
Tc = [1/2, 1, 2]
kb = 1.38064852e-23 # Boltzmann constant
hbar = 1.0545718e-34 # reduced Planck constant
h = 6.62607004e-34 # Planck constant
beta = lambda T: 1/(kb*T)
```

$tanh(\frac{zJm}{k_BT})$

$tanh(\frac{k_BT_Cm}{k_BT})$

with $T = [2,1,1/2]T_C$

$tanh(\frac{k_BT_Cm}{k_BaT_C})$

$tanh(\frac{m}{a})$ with $a = [2,1,1/2]$

```python
f1 = lambda a,m: np.tanh(m/a)
f2 = lambda m: m

m = np.linspace(-1, 1, 1000)

fig, ax = plt.subplots(1,3, figsize=(10, 4))

count = 0
for a in [2, 1, 1/2]:
    ax[count].plot(m, f1(a,m))
    ax[count].plot(m, f2(m))
    ax[count].set_xlabel('m')
    ax[count].set_ylabel('f(m)')
    ax[count].set_title(f'T/Tc = {a}')
    ax[count].legend(["tanh(m)", "m"])
    count += 1

fig.tight_layout()

```

b) Solve numerically (again by a Newton-Raphson method for example) Eq.(56) for temperatures
from $T/T_C = 2$ to $T/T_C = 0$ to show the onset of spontaneous magnetization

```python
f = lambda a,m: np.tanh(m/a) - m
df = lambda a,m: (1/a)*(1-np.tanh(m/a)**2) - 1

a = np.linspace(2, 0, 1000)

fig, ax = plt.subplots(1,1, figsize=(5, 5))
for i in a:
    f = lambda m: np.tanh(m/i) - m
    df = lambda m: (1/i)*(1-np.tanh(m/i)**2) - 1
    sol = newton(f, 1, df, tol=1e-5)

    ax.scatter(i, sol, c='r', s=1)
    ax.set_xlabel('T/Tc')
    ax.set_ylabel('m')
    ax.set_title('MAGNETIZATION vs T/Tc')
fig.tight_layout()
    
```

c) We also showed that the free energy has the form

$f(m; T) = -k_BT \ln [2 \cosh (\beta Jzm)]$ (57)

Plot $\Delta f(m; T) = f(m; T) - f(0; T)$ as a function of m for $T/T_C = [2, 1, 1/2]$ again in
the range $-1 <= m <= 1$ to show again the symmetry breaking process.

$f(m; T) = -k_BT \ln [2 \cosh (\beta Jzm)]$

$f(m; T) = -k_BT \ln [2 \cosh (\frac{Jzm}{k_BT})]$

$f(m; T) = -k_BT \ln [2 \cosh (\frac{k_BT_Cm}{k_BT})]$

with $T = [2,1,1/2]TC$

$f(m; T) = -k_BaT_C \ln [2 \cosh (\frac{k_BT_Cm}{k_BaT_C})]$

$f(m; T) = -k_BaT_C \ln [2 \cosh (\frac{m}{a})]$ with $a = [2,1,1/2]$


```python
m = np.linspace(-2, 2, 1000)

f = lambda m,a: m**2/2 - a*np.log(2*np.cosh(m/a))

fig, ax = plt.subplots(1,3, figsize=(10, 5))
count = 0
for a in [2, 1, 1/2]:
    ax[count].plot(m, f(m,a)-f(0,a))
    ax[count].set_xlabel('m')
    ax[count].set_ylabel(r'$\Delta f(m,T)$')
    ax[count].set_title(f'T/Tc = {a}')
    count += 1

fig.suptitle('FREE ENERGY vs MAGNETIZATION', fontsize=16)

fig.tight_layout()
```

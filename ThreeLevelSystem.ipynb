```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kb = 1.38064852e-23 # Boltzmann constant
hbar = 1.0545718e-34 # reduced Planck constant
h = 6.62607004e-34 # Planck constant
e = 1.60217662e-19 # elementary charge
eps0 = 0.1*e
N = 10000
V = 1
T = np.linspace(0.01, 100000, 1000000)
```

```python
def cv(T):
    b = 1/(kb*T)
    return (N*np.exp(9*b*eps0)*eps0**2)/(V*kb*T**2) *\
           (100*np.exp(b*eps0)+np.exp(10*b*eps0)+81)/\
           (1+np.exp(9*b*eps0)+np.exp(10*b*eps0))**2
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(T, cv(T))
ax[0].set_xlabel('Temperature (K)')
ax[0].set_ylabel('Heat Capacity (J/K)')
#ax[0].set_title('Heat Capacity of a Three-Level System')
ax[0].grid()

ax[1].plot(T, cv(T))
ax[1].set_xlabel('Temperature (K)')
ax[1].set_ylabel('Heat Capacity (J/K)')
#ax[1].set_title('Heat Capacity of a Three-Level System')
# set x axis in log scale
ax[1].set_xscale('log')
ax[1].grid()

fig.suptitle('Heat Capacity of a Three-Level System')

fig.tight_layout()
```

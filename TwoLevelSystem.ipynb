```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
kb = 8.617e-5  # eV/K
N = 1000

eps = 1.e-20
E = np.linspace(-N*eps, N*eps, int(1.e3))
```

```python
def entropy(E):
    return N*kb*np.log(2*N) - 0.5*N*kb*np.log(N**2 - (E/eps)**2)\
          - E/(2*eps)*kb*np.log((N+E/eps)/(N-E/eps))
```

```python
fig, ax = plt.subplots()



max_S = entropy(0)

ax.plot(E, entropy(E), c='b')
ax.hlines(max_S, -N*eps, N*eps, linestyle='--', color='g')
ax.vlines(0, 0, max_S, linestyle='--', color='g')
ax.set_title("2 levels system - Entropy S(E)", size=20)
ax.set_xlabel("Energy")
ax.set_ylabel("Entropy")
#ax.set_xscale('symlog')
ax.grid()
fig.tight_layout()
```

```python
def temperature(E):
    return -(2*eps)/kb * 1/np.log((N+E/eps)/(N-E/eps))
```

```python
fig, ax = plt.subplots()
ax.plot(E, temperature(E), c='b')
ax.set_title("2 levels system - Temperature T(E)", size=20)
ax.set_xlabel("Energy")
ax.set_ylabel("Temperature")
#ax.set_xscale('symlog')
ax.grid()
fig.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v0 = 1.0
v = np.linspace(-10, 10, 10000)
```

```python
# I case

def p1(v, v0):
    return np.exp(-abs(v)/v0) / (2*v0)


fig, ax = plt.subplots()
ax.plot(v, p1(v, v0))
ax.set_xlabel('v (m/s)')
ax.set_ylabel('p(v)')
ax.grid()
ax.set_title('I case')

```

```python
# II case

def p2(v, v0):
    return np.exp(-v**2/(2*v0**2)) / np.sqrt(2*np.pi*v0**2)


fig, ax = plt.subplots()
ax.plot(v, p2(v, v0))
ax.set_xlabel('v (m/s)')
ax.set_ylabel('p(v)')
ax.grid()
ax.set_title('I case')
```

```python
fig, ax = plt.subplots(2,2)

for v0 in [0.5, 1.0, 2.0, 5.0, 10]:
    ax[0,0].plot(v, p1(v, v0))
    ax[0,1].plot(v, p2(v, v0))

ax[0,0].set_xlabel('v (m/s)')
ax[0,0].set_ylabel('p(v)')
ax[0,0].grid()
ax[0,0].legend(['v0=0.5', 'v0=1.0', 'v0=2.0', 'v0=5.0', 'v0=10'])
ax[0,0].set_title('I case')

ax[0,1].set_xlabel('v (m/s)')
ax[0,1].set_ylabel('p(v)')
ax[0,1].grid()
ax[0,1].legend(['v0=0.5', 'v0=1.0', 'v0=2.0', 'v0=5.0', 'v0=10'])
ax[0,1].set_title('II case')


v0 = 1.0
ax[1,0].plot(v, p1(v, v0), c='k')
ax[1,0].plot(v, p2(v, v0), c='b')
ax[1,0].set_xlabel('v (m/s)')
ax[1,0].set_ylabel('p(v)')
ax[1,0].grid()
ax[1,0].legend(['I case', 'II case'])
ax[1,0].set_title('velocity distribution MAX ENT')

fig.tight_layout()
```
